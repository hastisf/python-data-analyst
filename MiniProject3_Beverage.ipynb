# Data Cleaning

## Cek Missing Value

```python
df.isnull()
```

```python
df.isnull().sum()
```

## Merapikan Data

```python
# Merapikan Data untuk Memudahkan Cleaning
df['Drink_Category'] = (
    df['Drink_Category']
    .str.strip()
    .str.title())

df['Membership_Level'] = (
    df['Membership_Level']
    .str.strip()
    .str.title())

df['Region'] = (
    df['Region']
    .str.strip()
    .str.title())
```

## Handle Missing Value

```python
# Isi Promo Pct kosong dengan 0
df['Promo_Pct'] = df['Promo_Pct'].fillna(0.0)
```

```python
# Isi Complaint Flag Berdasarkan Modus Complaint Membership
df['Complaint_Flag'] = df['Complaint_Flag'].fillna(
    df.groupby('Membership_Level')['Complaint_Flag'].transform(
        lambda x: x.mode()[0] if not x.mode().empty else np.nan))

df['Complaint_Flag'] = df['Complaint_Flag'].fillna(
    df['Complaint_Flag'].mode()[0])

df['Complaint_Flag'] = df['Complaint_Flag'].str.upper()

print(df['Complaint_Flag'].isna().sum())
```

```python
df['Complaint_Flag'].value_counts()
```

```python
# Missing value pada Membership Level Diisi Menggunakan Modus dari Cups Sold
df["Membership_Level"] = df["Membership_Level"].str.capitalize()
df["Membership_Level"] = df["Membership_Level"].fillna(
    df.groupby("Cups_Sold")["Membership_Level"].transform(
        lambda x: x.mode()[0] if not x.mode().empty else np.nan))
```

```python
df.head()
```

## Duplicate Checking

```python
df.duplicated().sum()
```

## Data Type Validation

```python
# Ubah Tipe Data Purchase_Date yang tidak beraturan dari Object ke Date Time - Mixed
df['Purchase_Date'] = pd.to_datetime(df['Purchase_Date'], format="mixed")

# Ubah Semua Tahun 2026 pada Data dengan 2025
df['Purchase_Date'] = df['Purchase_Date'].apply(
    lambda x: x.replace(year=2025) if x.year == 2026 else x
)

df.head()
```

## Data Consistency Checking

```python
# Melihat Daftar Kota yang Terdapat pada Kolom Store_City
df['Store_City'].unique()
```

```python
# Mengubah Inconsistency pada Store_City dengan Mapping - Replace
city_mapping= {
    'Jakrta': 'Jakarta',
    'Sby': 'Surabaya',
    'Bandng': 'Bandung',
    'Jogja': 'Yogyakarta'}
df['Store_City'] = df['Store_City'].replace(city_mapping)

# Cek Apakah Sudah Berubah
df['Store_City'].unique()
```

```python
df.head()
```

```python
# Melihat Daftar Kota yang Terdapat pada Kolom Region
df['Region'].unique()
```

```python
# Mengubah Inconsistency pada Region dengan Mapping - Ambil Data Store_City - Replace
region_baru = {
    'Yogyakarta': 'DIY',
    'Bandung': 'Jawa Barat',
    'Jakarta': 'DKI Jakarta',
    'Surabaya': 'Jawa Timur'}
df['Region'] = df['Store_City'].map(region_baru)

df.head()
```

```python
# Cek Apakah Sudah Berubah
df['Region'].unique()
```

# Feature Engineering

```python
# Tambah Kolom Gross Profit
df['Gross_Profit'] = df['Sales_Value'] - df['Ingredient_Cost']
df
```

```python
# Tambah Kolom Profit Margin Percentage
df['Profit_Margin_Pct'] = ((df['Gross_Profit'] / df['Sales_Value']) * 100).round(2)
df
```

```python
# Tambah Kolom Is Promo pada Promo_Pct > 0 dengan 1
df['Is_Promo'] = (df['Promo_Pct'] > 0).astype(int)
df
```

```python
# Tambah Kolom Klasifikasi Generasi Customer Berdasarkan Umur
def kategori_generasi(umur):
    if umur >= 14 and umur <= 29:
        return 'Gen Z'
    elif umur >= 30 and umur <= 45:
        return 'Millennial'
    elif umur >= 46 and umur <= 61:
        return 'Gen X'
    elif umur >= 62 and umur <= 80:
        return 'Boomer'
    else:
        return 'Lainnya'
df['Generasi'] = df['Customer_Age'].apply(kategori_generasi)
df
```

```python
# Tambah Kolom Is_Target_Achieved jika Achieved = 1
df['Is_Target_Achieved'] = np.where(df['Sales_Value'] >= df['Target_Sales'], 1, 0)
df
```

```python
# Menghitung Jumlah Transaksi yang Mencapai Target
df['Is_Target_Achieved'].value_counts()
```
