#Data Cleaning

## Salin Dataset

```python
# Salin Dataset untuk mempertahankan Data Asli

clean_df = df.copy()
```

## Cek Missing Values

```python
# Cek Apakah Ada Missing Values

clean_df.isnull()
```

```python
# Cek Jumlah Missing Value

clean_df.isnull().sum()
```

## Handle Missing Value

```python
# Isi children kosong dengan modus karena mean = modus = 1
clean_df['children'] = clean_df['children'].fillna(
    clean_df['country'].mode()[0])

# Isi country kosong dengan negara terbanyak karena tipe data object
clean_df['country'] = clean_df['country'].fillna(
    clean_df['country'].mode()[0])

# Isi agent kosong dengan 0
clean_df['agent'] = clean_df['agent'].fillna(0)

# Drop kolom company karena >94% data berisi NULL
clean_df.drop(['company'], axis=1, inplace=True)

# Isi meal undefined dengan modus
clean_df['meal'] = clean_df['meal'].replace(
    'Undefined', 'unknown')
```

## Hapus Angka Tidak Valid




```python
# Menghapus data adr yang negatif
clean_df = clean_df[clean_df['adr'] >= 0]

# Menghapus jumlah tamu adults negatif
clean_df = clean_df[
    clean_df['adults'] >= 0]

# Menghapus jumlah tamu children negatif
# clean_df = clean_df[clean_df['children'] >= 0]
# run kembali setelah ubah data type

# Menghapus jumlah tamu babies negatif
clean_df = clean_df[
    clean_df['babies'] >= 0]
```

## Hapus Outlier ADR


```python
# Mengambil nilai persentil 99 dari kolom adr
q99 = clean_df['adr'].quantile(0.99)

# Menghapus nilai adr yang terlalu tinggi (outlier 1%)
clean_df = clean_df[
    clean_df['adr'] <= q99]
```

## Ubah Tipe Data

```python
# Melihat Tipe Data Awal

clean_df.info()
```

```python
# Mengubah Tipe Data dari float64 ke int64

# clean_df['children'] = clean_df['children'].fillna(0).astype('int64')
# gagal run karena mungkin ada data NaN atau tanpa angka

# data aneh/non-angka otomatis dijadikan NaN -> diisi 0 -> int64
clean_df['children'] = pd.to_numeric(clean_df['children'],
    errors='coerce').fillna(0).astype('int64')

# Mengubah Tipe Data dari float64 ke int64 agar konsisten
clean_df['agent'] = clean_df['agent'].fillna(0).astype('int64')

# Mengubah Tipe Data dari Object ke Datetime
clean_df['reservation_status_date'] = pd.to_datetime(
    clean_df['reservation_status_date'])
```

```python
# Melihat Data setelah Diubah Tipe Data

clean_df.info()
```

```python
# Menghapus jumlah tamu children negatif

clean_df = clean_df[clean_df['children'] >= 0]
```

## Merapikan Data

```python
# Menghilangkan Spasi Berlebih
# Mengganti Spasi Antar Kata dengan _
# Data Dibuat Lower

clean_df['hotel'] = clean_df['hotel'].str.strip().str.lower().str.replace(' ', '_')
clean_df['arrival_date_month'] = clean_df['arrival_date_month'].str.strip().str.lower().str.replace(' ', '_')
clean_df['meal'] = clean_df['meal'].str.strip().str.lower().str.replace(' ', '_')
clean_df['country'] = clean_df['country'].str.strip().str.lower().str.replace(' ', '_')
clean_df['market_segment'] = clean_df['market_segment'].str.strip().str.lower().str.replace(' ', '_')
clean_df['distribution_channel'] = clean_df['distribution_channel'].str.strip().str.lower().str.replace(' ', '_')
clean_df['reserved_room_type'] = clean_df['reserved_room_type'].str.strip().str.lower().str.replace(' ', '_')
clean_df['assigned_room_type'] = clean_df['assigned_room_type'].str.strip().str.lower().str.replace(' ', '_')
clean_df['deposit_type'] = clean_df['deposit_type'].str.strip().str.lower().str.replace(' ', '_')
clean_df['customer_type'] = clean_df['customer_type'].str.strip().str.lower().str.replace(' ', '_')
clean_df['reservation_status'] = clean_df['reservation_status'].str.strip().str.lower().str.replace(' ', '_')
```

# Klasifikasi ADR

```python
clean_df['adr']
```

```python
# Mengetahui Sebaran Nilai ADR

pd.DataFrame({
'max' : [clean_df['adr'].max()],
'min' : [clean_df['adr'].min()],
'mode' : [clean_df['adr'].mode()[0]],
'median' : [clean_df['adr'].median()],
'mean' : [clean_df['adr'].mean()]
})
```

```python
# Agar Nilai Absolut

clean_df['adr'] = clean_df['adr'].abs()

#Klasifikasi ADR

def adrcategory(adr):

  if adr == 0:
    return 'free'

  elif 1 <= adr <= 75:
    return 'low'

  elif 76 <= adr <= 150:
    return 'high'

  else:
    return 'premium'
```

# Membuat Kolom Baru adr_category

```python
# Membuat Kolom adr_category Berdasarkan adr

clean_df['adr_category'] = (df['adr'].apply(adrcategory))
```

#Insight


## 1. Pembatalan Berdasarkan Tipe Deposit



```python
total_pesanan = clean_df.groupby('deposit_type')['is_canceled'].count()
total_batal = clean_df.groupby('deposit_type')['is_canceled'].sum()
persen_batal = (total_batal/total_pesanan*100).round(2)

tabel_deposit = pd.DataFrame({
    'total_pesanan': total_pesanan,
    'total_batal': total_batal,
    'persen_batal': persen_batal})

tabel_deposit = tabel_deposit.sort_values(by='persen_batal', ascending=False)

print(tabel_deposit)
```

## 2. Total Tamu Menginap Berdasarkan arrival_date_month


```python
total_tamu = (clean_df[clean_df['is_canceled'] == 0].groupby
 (['hotel','arrival_date_month'])['is_canceled'].count())

rata_rata_harga_kamar = (clean_df[clean_df['is_canceled']==0].groupby
      (['hotel','arrival_date_month'])['adr'].mean().round(2))

total_tamu_by_month = pd.DataFrame({
    'total_tamu': total_tamu,
    'rata_rata_harga_kamar': rata_rata_harga_kamar})

total_tamu_by_month = total_tamu_by_month.sort_values(
    by='rata_rata_harga_kamar', ascending=False).head(20)

print(total_tamu_by_month)
```

## 3. Top 5 Rank Tamu Berdasarkan Negara Asal

```python
asal_negara = (
    clean_df[clean_df['is_canceled'] == 0]
    .groupby('country')['is_canceled']
    .count())

top_rank_negara = pd.DataFrame({
    'total_tamu': asal_negara
})

top_rank_negara = top_rank_negara.sort_values(by='total_tamu',
ascending=False).head(5)

print(top_rank_negara)
```

## 4. Total Special Request Berdasarkan Market Segment

```python
special_request = (clean_df[clean_df['is_canceled'] ==0].groupby
 ('market_segment')['total_of_special_requests'].mean().round(2))

top_market_segment = pd.DataFrame({
    'rata_rata_special_request': special_request})

top_market_segment= top_market_segment.sort_values(by='rata_rata_special_request',
ascending=False).head()

print(top_market_segment)
```

## 5. Jumlah Pesanan Berdasarkan Perubahan Pesanan

```python
booking_changes = (clean_df.groupby('booking_changes').size().index)

total_pesanan = (clean_df.groupby('booking_changes')['is_canceled'].size())

total_batal = (clean_df.groupby('booking_changes')['is_canceled'].sum())

summary = pd.DataFrame({
    'booking_changes': booking_changes,
    'total_pesanan': total_pesanan.values,
    'total_batal': total_batal.values})

summary = summary.sort_values(
    by='booking_changes',
    ascending=True).head()

print(summary)
```

## 6. Jumlah Tipe Pesanan Makanan Berdasarkan Hotel

```python
tipe_paket_makan = (clean_df[clean_df['is_canceled'] == 0].groupby
 (['hotel','meal'])['is_canceled'].count())

total_meal_dipesan = pd.DataFrame({
    'total_dipesan': tipe_paket_makan})

top_meal_dipesan = total_meal_dipesan.sort_values(
    by=['hotel', 'total_dipesan'],
    ascending=[True, False])

print(top_meal_dipesan)
```

## 7. Total Pesanan Berdasarkan Agent

```python
total_agen_pesan = (clean_df[clean_df['agent'] !=0].groupby('agent')
['is_canceled'].count())

total_agen_batal = (clean_df[clean_df['agent'] !=0].groupby('agent')
['is_canceled'].sum())

agen_batal = pd.DataFrame({
    'total_agen_pesan': total_agen_pesan,
    'total_agen_batal': total_agen_batal})

agen_batal = agen_batal.sort_values(by='total_agen_batal',
    ascending=False).head()

print(agen_batal)
```
